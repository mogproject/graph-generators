## Random Near-modular Graphs

```python
from random import Random
from math import *
import networkx as nx
from generator.random_near_modular_graph import random_near_modular_graph
```

```python
def draw(G, k):
    angle_rand = Random(123)
    angles = [pi / 2]
    for i in range(k - 1):
        angles += [angles[-1] + 2 * pi / k]

    level = 1
    while k ** level < len(G):
        level += 1

    def f(config):
        cf = [c + pi / 40 * (angle_rand.random() - 0.5) for c in config]
        ts = [3 ** i + (0.5 + angle_rand.random()) * 0.4 * i for i in range(len(cf))]
        x = sum(t * cos(c) for t, c in zip(ts, cf))
        y = sum(t * sin(c) for t, c in zip(ts, cf))
        return x, y

    pos = {}
    for i in range(len(G)):
        config = []
        t = i
        for j in range(level):
            config += [angles[t % k]]
            t //= k
        pos[i] = f(config)

    nx.draw(G, node_size=50, alpha=0.7, pos=pos)
```
